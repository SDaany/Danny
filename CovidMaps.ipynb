```
import numpy as np
import pandas as pd 
import csv
import folium
import datetime
from folium import plugins
from folium.plugins import HeatMap, MarkerCluster
!wget --no-check-certificate \
   https://covid.ourworldindata.org/data/owid-covid-data.csv \
    -O /data.csv
!wget --no-check-certificate \
   https://docs.google.com/spreadsheets/d/e/2PACX-1vSSd4CA6Bq88H4_GdKuG8f5u7yrASn58gklL61yCexF3XEmMTxgKbs-0PbKzTNINef55Rq4Yem5CE7l/pub?output=csv \
    -O /geo_data.csv  
d=pd.read_csv("/data.csv")
g=d.groupby("date")
c=d.groupby("location")
dates=[ date for date , date_df in g]
ctries=[str(i) for i , df in c]
dte=dates[-2]    
cases=g.get_group(dte)  
l=pd.read_csv("/geo_data.csv", error_bad_lines=False)
l2=set(l.name)
l1=set(ctries)
data0=[[float(l.loc[l.name==i]["latitude"]),float(l.loc[l.name==i]["longitude"]),(cases.loc[cases.location==i]["new_cases"])] for i in l1.intersection(l2)]
for j in data0:
  if len(list(j[2]))==0:
    j[2]=0
  else:
    j[2]=int(j[2]) 
```

**NEW CASES**

```


# map
map_enum_icons = folium.Map([50, 30], zoom_start=5)

for i in data0:
    folium.Marker(location=[i[0], i[1]],
                  icon=plugins.BeautifyIcon(
                  number=int(i[2]),
                  border_color='blue',
                  border_width=1,
                  text_color='black',
                  )).add_to(map_enum_icons)

# display map    
map_enum_icons
```

```
map=folium.Map(location=[50,30], zoom_start=5)
HeatMap(data0,radius=25).add_to(map)
map
```
